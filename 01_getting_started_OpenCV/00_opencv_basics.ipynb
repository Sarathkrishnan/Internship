```python
import cv2
import matplotlib.pyplot as plt
plt
#reading image
path='watch.jpg'
image = cv2.imread(path)
plt.imshow(image)
```

```python
# extracting the height and width of an image
h,w = image.shape[:2]
print("Height = {}, Width ={}".format(h,w))
```

```python
#extracting RGB values
(B,G,R) = image[100,100]
print('R = {}, G = {}, B = {}'.format(B,G,R))

#extracting value from specific channel
B = image[100,100,0]
print("B = {}".format(B))
```

```python
#extracting the region of interest(ROI)
roi = image[0:500, 0:600]
```

```python
# Resize the image
resize = cv2.resize(image,(800, 800))
```

```python
# Resize without compromizing aspect ratio

#1. calculating ratio
ratio = 800 / w
#2. create tuple contains width and height
dim = (800, int(h * ratio))
#3. Resizing image
resize_aspect = cv2.resize(image,dim)
plt.imshow(resize_aspect) #plt not affect in case of resize
```

```python
# Rotating the image
center = (w//2,h//2)

#generating a rotation matrix
matrix = cv2.getRotationMatrix2D(center, -45, 1.0)

# performing the affine transformation
rotated = cv2.warpAffine(image, matrix, (w,h))

plt.imshow(rotated)
```
